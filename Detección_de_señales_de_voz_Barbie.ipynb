# Procesar las bases de datos

```python
from pydub import AudioSegment
import os
import glob

def procesar_dataset(ruta_entrada, ruta_salida, duracion_segundos=2):
    duracion_ms = duracion_segundos * 1000

    if not os.path.exists(ruta_salida):
        os.makedirs(ruta_salida)

    archivos = glob.glob(f"{ruta_entrada}/**/*.wav", recursive=True) + \
               glob.glob(f"{ruta_entrada}/**/*.mp3", recursive=True) + \
               glob.glob(f"{ruta_entrada}/**/*.flac", recursive=True)

    print(f"Se encontraron {len(archivos)} archivos. Procesando...")

    clips_generados = 0
    for archivo in archivos:
        try:
            audio = AudioSegment.from_file(archivo)
            nombre_base = os.path.basename(archivo).split('.')[0]

            for i, chunk in enumerate(audio[::duracion_ms]):
                if len(chunk) == duracion_ms:
                    nombre_salida = f"{nombre_base}_clip_{i}.wav"
                    ruta_guardado = os.path.join(ruta_salida, nombre_salida)
                    chunk.export(ruta_guardado, format="wav")
                    clips_generados += 1
        except Exception as e:
            # Se ignoran los archivos corruptos
            pass

    print(f"Se generaron {clips_generados} clips de {duracion_segundos} segundos exactos.")
```

```python
ruta_chm150 = "/content/drive/MyDrive/Proyecto_Voz/CHM150"
ruta_tedx = "/content/drive/MyDrive/Proyecto_Voz/tedx_spanish_corpus"

ruta_salida_dataset = "/content/drive/MyDrive/Proyecto_Voz/Dataset_2s"

print("Recorte de CHM150 a 2 segundos")
procesar_dataset(ruta_entrada=ruta_chm150,
                       ruta_salida=f"{ruta_salida_dataset}/CHM150",
                       duracion_segundos=2)

print("\n Recorte de TEDx a 2 segundos")
procesar_dataset(ruta_entrada=ruta_tedx,
                       ruta_salida=f"{ruta_salida_dataset}/TEDx",
                       duracion_segundos=2)
```
